```python
lst1 = []
lst2 = []
```

```python
for i in range(repeat):
    lst1.append(f"https://books.toscrape.com/catalogue/page-{i+1}.html")
```

```python
lst1
```

```python
def convert(num):
    values = {
        'One':1,
        'Two':2,
        'Three':3,
        'Four':4,
        'Five':5
    }

    return values[num]
```

```python
for link in lst1:
    response = requests.get(link)
    soup = BeautifulSoup(response.text)
    articles = soup.find_all('article', class_='product_pod')
    art = articles[0]

    for art in articles:

        #Part 1: Title
        h3 = art.find('h3')
        a = h3.find('a')

        title = a.attrs['title']

        #Part 2: Price
        price = art.find('p', class_="price_color").text
        price = float(price[2:])

        #Part 3: Rating
        rating = art.find('p', class_="star-rating")
        rating = rating.attrs['class'][-1]

        num_rating = convert(rating)

        #Part 4: Saving Data
        book_data = {'Title': title, 'Price': price, 'Rating': num_rating}
        lst2.append(book_data)

        #Part 5: Printing Results
        print(f"Book Data: {book_data}")
```

```python
with open ('Task 3.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['Title','Price','Rating'])
    writer.writeheader()
    
    for book in lst2:
        writer.writerow(book)
```
